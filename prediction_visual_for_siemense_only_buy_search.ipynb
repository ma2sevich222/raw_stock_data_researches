```python
import pandas as pd
import plotly.express as px
import numpy as np
import plotly.graph_objects as go
from collections import Counter
```

```python
DATA=pd.read_csv('pattern_model_test.csv')
```

```python
DATA=DATA.rename(columns={"pattern No.":"pattern"})
```

```python
def ploter(data,treshhold,patt):

  

   Demo=data[(data.distance <=treshhold)&(DATA.signal == 1)&(DATA.pattern==patt)]


  

   fig1 = px.line(x=data['date'],y=data['close'])

   fig2=px.scatter(x=Demo['date'],y=Demo['close'],color_discrete_sequence=['green'],title="Buy Signals")
   fig2.update_traces(marker_size=10)
   fig2.update_xaxes(type='category')
   fig = go.Figure(data=fig1.data + fig2.data)
   fig.update_xaxes(title_text="DATE")
   fig.update_yaxes(title_text="CLOSE")
   fig.show()

  
```

#Точки сигналов BUY на временном ряде CLOSE

```python
ploter(DATA,0.03,59)
```
